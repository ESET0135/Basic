```python
import numpy as np
import pandas as pd
```

```python
arr=np.array([10,20,30])
df=pd.DataFrame({'num':arr})
plt.bar(df.index,df["num"])
plt.show()
```

```python
data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [24, 30, 35, 40],
    "Salary": [50000, 60000, 75000, 1000000]
}

df = pd.DataFrame(data)
print(df)
plt.figure(figsize=(6,4))
plt.bar(df["Name"], df["Salary"])
plt.xlabel("Name")
plt.ylabel("Salary")
plt.title("Salary by Employee")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(8, 6))
sns.boxplot(y=tips_data['total_bill'])
plt.title('Box Plot of Total Bill Amount')
plt.ylabel('Total Bill ($)')
plt.show()


```
